#### Resposta Questão 1 Página 95
Para localizar graficamente as raízes, precisamos definir um intervalo de plotagem no qual o gráfico será construído. Como estamos utilizando o matplotlib em Python, não temos uma ferramenta de visualização automática: precisamos fornecer ao programa os valores de $x$ em que a função será avaliada. A escolha adequada do intervalo é fundamental, pois garante que a curva seja desenhada em uma região onde ocorre a mudança de sinal da função, permitindo visualizar as raízes (conforme o Teorema do Valor Intermediário).


**d) $2^x - 3x = 0$**

Vamos analisar alguns valores:

- Para $x = 0$:
$$2^0 - 3 \cdot 0 = 1$$
$$ 1 > 0$$

- Para $x = 1$:

$$2^1 - 3 \cdot 1 = 2 - 3 = -1$$
$$ -1 < 0$$

Portanto, há uma mudança de sinal entre $x = 0$ e $x = 1$, indicando uma raiz nesse intervalo.

Analisando valores maiores:

- Para $x = 3$: 
$$2^3 - 3 \cdot 3 = 8 - 9 = -1 $$
$$-1 < 0$$

- Para $x = 4$:
$$2^4 - 3 \cdot 4 = 16 - 12 = 4 $$
$$4 > 0$$

Aqui também ocorre uma mudança de sinal entre $x = 3$ e $x = 4$, indicando uma segunda raiz.

Assim, a função $2^x - 3x$ possui duas raízes reais, uma entre $x = 0$ e $x = 1$, e outra entre $x = 3$ e $x = 4$.

**e) $x^3 + x - 1000 = 0$**

Para estimar a raiz, note que $x^3$ domina para valores grandes de $x$. Se $x = 10$, temos:

$10^3 + 10 - 1000 = 1000 + 10 - 1000 = 10 > 0$

Se $x = 9$, temos:

$9^3 + 9 - 1000 = 729 + 9 - 1000 = -262 < 0$

Portanto, há uma mudança de sinal entre $x = 9$ e $x = 10$, indicando que a raiz está nesse intervalo.

Essas análises justificam os intervalos iniciais escolhidos para a visualização gráfica das raízes. Que foram $[0,4]$ para a letra $d)$ e $[9,10]$ para a letra $e)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define as funções
def f1(x):
    return 2**x - 3*x

def f2(x):
    return x**3 + x - 1000

# Intervalos iniciais para visualização geral
x_vals1 = np.linspace(0, 4, 400)      # Para f1(x) = 2^x - 3x
x_vals2 = np.linspace(9, 10, 400)     # Para f2(x) = x^3 + x - 1000

# Zoom nos intervalos das raízes de f1
x_zoom1a = np.linspace(0.4, 0.6, 400)     # Primeira raiz de f1
x_zoom1b = np.linspace(3, 3.6, 400)     # Segunda raiz de f1

# Zoom no intervalo da raiz de f2
x_zoom2 = np.linspace(9.9, 10, 400)     # Raiz de f2

fig, axs = plt.subplots(3, 2, figsize=(14, 12))

# Gráfico geral de f1
axs[0, 0].plot(x_vals1, f1(x_vals1), label='$2^x - 3x$')
axs[0, 0].axhline(0, color='black', linewidth=0.8)
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')
axs[0, 0].set_title('Visão Geral: $2^x - 3x = 0$')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Gráfico geral de f2
axs[0, 1].plot(x_vals2, f2(x_vals2), label='$x^3 + x - 1000$')
axs[0, 1].axhline(0, color='black', linewidth=0.8)
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')
axs[0, 1].set_title('Visão Geral: $x^3 + x - 1000 = 0$')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Zoom na primeira raiz de f1
axs[1, 0].plot(x_zoom1a, f1(x_zoom1a), label='$2^x - 3x$ (zoom raiz 1)')
axs[1, 0].axhline(0, color='black', linewidth=0.8)
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')
axs[1, 0].set_title('Zoom: $2^x - 3x = 0$ (primeira raiz)')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Zoom na segunda raiz de f1
axs[2, 0].plot(x_zoom1b, f1(x_zoom1b), label='$2^x - 3x$ (zoom raiz 2)')
axs[2, 0].axhline(0, color='black', linewidth=0.8)
axs[2, 0].set_xlabel('x')
axs[2, 0].set_ylabel('f(x)')
axs[2, 0].set_title('Zoom: $2^x - 3x = 0$ (segunda raiz)')
axs[2, 0].legend()
axs[2, 0].grid(True)

# Zoom na raiz de f2
axs[1, 1].plot(x_zoom2, f2(x_zoom2), label='$x^3 + x - 1000$ (zoom)')
axs[1, 1].axhline(0, color='black', linewidth=0.8)
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')
axs[1, 1].set_title('Zoom: $x^3 + x - 1000 = 0$')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Remove o subplot não utilizado (inferior direito)
fig.delaxes(axs[2, 1])

plt.tight_layout()
plt.show()
```

#### Questão 2 Página 96

O método da bissecção pode ser aplicado sempre que $f(a) \times f(b) \lt 0$, mesmo que $f(x)$ tenha mais que um zero em (a, b). Nos casos em que isso ocorre, verifique com o auxílio de gráficos, se é possível determinar qual zero será obtido por este método.

```python
import numpy as np
import matplotlib.pyplot as plt

# Função com duas raízes
def real_function(x):
    return (x - 1) * (x - 2) * (x - 3)

def bisection_steps(f, a, b, k):
    history = []
    for i in range(k):
        m = (a + b) / 2
        history.append((a, b, m))
        if f(m) == 0:
            break
        elif f(a) * f(m) < 0:
            b = m
        else:
            a = m
    return history

# Lista de intervalos para testar diferentes raízes
intervals = [
    (0, 4),   # Aproxima raiz perto de x=1
    (0.5, 4),   # Aproxima raiz perto de x=2
    (0, 3.5),   # Aproxima raiz perto de x=3
]

k = 6  # número de iterações

fig, axs = plt.subplots(1, len(intervals), figsize=(5 * len(intervals), 5))

for idx, (a, b) in enumerate(intervals):
    x = np.linspace(a, b, 400)
    axs[idx].plot(x, real_function(x), label='$f(x) = (x - 1)(x - 2)(x - 3)$')
    axs[idx].axhline(0, color='black', linewidth=0.8)
    history = bisection_steps(real_function, a, b, k)
    for i, (a_i, b_i, m_i) in enumerate(history):
        axs[idx].axvline(a_i, color='red', linestyle='--', alpha=0.7, label='a' if i == 0 else None)
        axs[idx].axvline(b_i, color='blue', linestyle='--', alpha=0.7, label='b' if i == 0 else None)
        axs[idx].axvline(m_i, color='green', linestyle=':', alpha=0.7, label='m' if i == 0 else None)
    axs[idx].set_title(f'Bisseção em [{a}, {b}]')
    axs[idx].set_xlabel('x')
    axs[idx].set_ylabel('f(x)')
    axs[idx].legend()
    axs[idx].grid(True)

plt.tight_layout()
plt.show()
```

#### Resposta Questão 2 Página 96

O método da bissecção exige apenas que $f(a)\,f(b)<0$. Se dentro do intervalo $(a,b)$ houver mais de uma raiz, o método **ainda** pode ser aplicado, porém tendo como informação inicial, somente a função e o método a ser utilizado **não existe garantia** de qual das raízes contidas em $(a,b)$ será encontrada. Essa foi a conclusão que chegamos ao realizar o experimento gráfico acima. Onde ao variar o valor inicial de $a$ e $b$ nos aproximamos de diferentes raizes.

Em resumo:
- o critério \(f(a)f(b)<0\) é suficiente para aplicar a bissecção mesmo com várias raízes no intervalo;
- **qual** raiz será encontrada depende do intervalo inicial — para direcionar o método a uma raiz específica o mais simples é escolher um intervalo que a isole;
- os gráficos ajudam a identificar subintervalos apropriados antes de aplicar o método numérico.

#### Questão 12 Página 97
Aplique o método de Newton-Raphson à equação:

$x^3 - 2x^2 -3x +10 = 0$ com $x_0 = 1.9$

Justifique o que acontece.

```python
import numpy as np
import pandas as pd

# Definições
def f(x):
    return x**3 - 2*x**2 - 3*x + 10

def df(x):
    return 3*x**2 - 4*x - 3

def newton_table(f, df, x0, tol=1e-6, max_iter=40):
    data = []
    xk = x0
    for k in range(max_iter):
        fxk = f(xk)
        dfxk = df(xk)
        if dfxk == 0:
            data.append([k, xk, fxk, dfxk, None, None])
            break
        xk1 = xk - fxk/dfxk
        error = abs(xk1 - xk)
        data.append([k, xk, fxk, dfxk, xk1, error])
        if error < tol:
            break
        xk = xk1
    return pd.DataFrame(data, columns=["k", "x_k", "f(x_k)", "f'(x_k)", "x_{k+1}", "|x_{k+1} - x_k|"])

# Execução
x0 = 1.9
df_newton = newton_table(f, df, x0)
display(df_newton.head(40))
```

##### Observação do Comportamento Inicial

Ao aplicar o **método de Newton-Raphson** para encontrar a raiz da equação $f(x) = x^3 - 2x^2 - 3x + 10 = 0$ com um chute inicial $x_0 = 1.9$, observamos que as primeiras iterações produzem **saltos enormes** nos valores de $x_k$, conforme mostrado na tabela de resultados.

A razão para esse comportamento é a fórmula de iteração do método, dada por:

$$x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)}$$

O problema ocorre quando o denominador, a derivada $f'(x_k)$, é muito próximo de zero. Nesse caso, a divisão $\frac{f(x_k)}{f'(x_k)}$ resulta em um valor muito grande, fazendo com que o próximo termo, $x_{k+1}$, salte para longe do valor atual $x_k$.

Para nosso problema, a derivada da função é $f'(x) = 3x^2 - 4x - 3$. No primeiro passo ($k=0$), temos:

* $x_0 = 1.9$
* $f(x_0) = f(1.9) = (1.9)^3 - 2(1.9)^2 - 3(1.9) + 10 = 3.939$
* $f'(x_0) = f'(1.9) = 3(1.9)^2 - 4(1.9) - 3 = 0.23$

A derivada em $x_0=1.9$ é **muito pequena** ($0.23$), o que leva a um passo de correção muito grande. A iteração para $x_1$ é:

$$x_1 = 1.9 - \frac{3.939}{0.23} \approx 1.9 - 17.126 = -15.226$$

Esse primeiro salto para um valor tão distante ($x_1 \approx -15.226$) faz com que as iterações subsequentes também apresentem grandes variações, pois os valores de $f(x_k)$ e $f'(x_k)$ para $x$ tão distantes da origem são numericamente grandes em magnitude. No entanto, após essas oscilações iniciais, o método converge para a raiz da função.

##### Convergência para a Raiz

Apesar dos saltos iniciais, o método eventualmente encontra a **bacia de atração** da raiz $x = -2$. O valor $-2$ é, de fato, uma raiz exata da equação original:

$$f(-2) = (-2)^3 - 2(-2)^2 - 3(-2) + 10 = -8 - 8 + 6 + 10 = 0$$

Quando as iterações se aproximam de $x=-2$, o método volta a se comportar de maneira estável e convergente, pois a derivada da função nesse ponto não é nula:

$$f'(-2) = 3(-2)^2 - 4(-2) - 3 = 12 + 8 - 3 = 17$$

Com um valor de derivada relativamente grande ($17$), os passos de iteração se tornam pequenos e estáveis, garantindo a rápida convergência para a raiz $-2$.

##### Justificativa do Comportamento

Em resumo, a escolha do chute inicial $x_0 = 1.9$ foi problemática porque o valor da derivada $f'(1.9)$ é muito próximo de zero. Isso gera um grande passo de iteração inicial, causando os saltos observados na tabela. No entanto, após algumas iterações erráticas, o método ainda veio a ser atraído para a raiz real da função, $x = -2$, onde a convergência ocorre de forma rápida e estável.

A lição prática é que a escolha de um chute inicial é crucial para a eficiência do método de Newton. Em casos onde a derivada pode ser nula ou muito pequena, o método pode falhar ou exigir muitas iterações para convergir, se convergir. Alternativas incluem a escolha de um chute inicial mais próximo da raiz ou o uso de outros métodos, como a bissecção, para encontrar uma boa estimativa inicial.
